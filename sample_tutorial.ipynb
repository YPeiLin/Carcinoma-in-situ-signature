# [4] Downstream analysis example: Association with survival

```r
library(survival)
rm(list=ls())
myinf1 = "/mount/amos1/home/ylin/sample_input/sample_GenomicEvent_iRAS_SQC.txt" # from section 3
myinf2 = "/mount/amos1/home/ylin/sample_input/sample_Clinical_info.txt" # sample clinical outcome

#data preprocess
info = read.table(myinf2, sep="\t", header=T, row.names=1, quote="")
t.surv = as.numeric(info$os_mo)
e.surv = info$os_event
info = cbind(t.surv, e.surv, info)
tag = !is.na(info$t.surv)
info = info[tag==1,]
dim(info)


data = read.table(myinf1, sep="\t", header=T, row.names=1)
rownames(data) = gsub("X", "", rownames(data))
cnum = ncol(data)/4
data = data[, 1:cnum] - data[, (1+cnum):(cnum*2)]
tmp = colnames(data)
tmp = gsub("\\.up\\.ES", "", tmp)
colnames(data) = tmp

## common samples
comxx = intersect(row.names(data), row.names(info))
data = data[comxx,]
info = info[comxx,]
data
```
